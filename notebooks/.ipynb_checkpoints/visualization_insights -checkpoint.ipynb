```python
import re
import warnings
from sklearn.feature_extraction.text import TfidfVectorizer
import seaborn as sns
import oracledb
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import WordCloud

# Suppress warnings
warnings.filterwarnings("ignore")
sns.set_style("whitegrid")

# Database connection details
hostname = 'localhost'
port = 1521
sid = 'bankreviews'
username = 'SYSTEM'
password = 'Opass#25'
dsn = oracledb.makedsn(hostname, port, sid=sid)

try:
    # Establish connection to Oracle Database
    connection = oracledb.connect(user=username, password=password, dsn=dsn)
    
    # SQL query to fetch reviews data
    query = """
    SELECT
        review_text,
        rating,
        review_date,
        bank_name,
        source,
        processed_review,
        sentiment_score,
        sentiment
    FROM
        reviews
    """
    
    # Load data into DataFrame
    df = pd.read_sql(query, connection)
    print(f"Total records loaded: {len(df)}")

except oracledb.Error as e:
    error_obj, = e.args
    print(f"Error connecting to Oracle or querying data: {error_obj.message}")
    print("Please ensure your database is running and connection details are correct (username, password, DSN).")
    exit()
finally:
    # Close connection if it was established
    if 'connection' in locals() and connection:
        connection.close()

# Data preprocessing
df['REVIEW_DATE'] = pd.to_datetime(df['REVIEW_DATE'], errors='coerce')
df['RATING'] = pd.to_numeric(df['RATING'], errors='coerce')
df['SENTIMENT_SCORE'] = pd.to_numeric(df['SENTIMENT_SCORE'], errors='coerce')

# Drop rows with missing values in important columns
df.dropna(subset=['RATING', 'SENTIMENT_SCORE', 'REVIEW_DATE', 'PROCESSED_REVIEW', 'SENTIMENT'], inplace=True)

# Ensure 'PROCESSED_REVIEW' and 'SENTIMENT' are in correct format
df['PROCESSED_REVIEW'] = df['PROCESSED_REVIEW'].astype(str).fillna('')
df['SENTIMENT'] = df['SENTIMENT'].astype(str).fillna('neutral').str.lower() 

# Display data types after loading and initial processing
print("\nData types after loading and initial processing for analysis:")
print(df.info())

# Identifying drivers (top positive keywords)
positive_reviews = df[df['SENTIMENT'] == 'positive']['PROCESSED_REVIEW']
negative_reviews = df[df['SENTIMENT'] == 'negative']['PROCESSED_REVIEW']

print("\n--- Identifying Drivers (Top Positive Keywords)")
drivers = {}
if not positive_reviews.empty and not positive_reviews.str.strip().eq('').all():
    non_empty_positive_text = positive_reviews[positive_reviews.str.strip() != '']
    if not non_empty_positive_text.empty:
        vectorizer_drivers = TfidfVectorizer(stop_words='english', ngram_range=(1, 2), max_features=1000)
        tfidf_matrix_drivers = vectorizer_drivers.fit_transform(non_empty_positive_text)
        feature_names_drivers = vectorizer_drivers.get_feature_names_out()
        sums_drivers = tfidf_matrix_drivers.sum(axis=0)
        scores_drivers = [(feature_names_drivers[col], sums_drivers[0, col]) for col in range(len(feature_names_drivers))]
        scores_drivers.sort(key=lambda x: x[1], reverse=True)
        drivers = {word: score for word, score in scores_drivers[:10]}

if drivers:
    for word, score in drivers.items():
        print(f"- {word} (Score: {score:.2f})")
else:
    print("No sufficient positive reviews or keywords identified.")
# Identifying pain points (top negative keywords)
print("\n--- Identifying Pain Points (Top Negative Keywords)")
pain_points = {}
if not negative_reviews.empty and not negative_reviews.str.strip().eq('').all():
    non_empty_negative_text = negative_reviews[negative_reviews.str.strip() != '']
    if not non_empty_negative_text.empty:
        vectorizer_pain_points = TfidfVectorizer(stop_words='english', ngram_range=(1, 2), max_features=1000)
        tfidf_matrix_pain_points = vectorizer_pain_points.fit_transform(non_empty_negative_text)
        feature_names_pain_points = vectorizer_pain_points.get_feature_names_out()
        sums_pain_points = tfidf_matrix_pain_points.sum(axis=0)
        scores_pain_points = [(feature_names_pain_points[col], sums_pain_points[0, col]) for col in range(len(feature_names_pain_points))]
        scores_pain_points.sort(key=lambda x: x[1], reverse=True)
        pain_points = {word: score for word, score in scores_pain_points[:10]}

if pain_points:
    for word, score in pain_points.items():
        print(f"- {word} (Score: {score:.2f})")
else:
    print("No negative reviews or keywords identified.")

# Comparing banks
print("\n--- Comparing Banks ---")
avg_rating_per_bank = df.groupby('BANK_NAME')['RATING'].mean().sort_values(ascending=False)
print("\nAverage Rating per Bank:")
print(avg_rating_per_bank)

sentiment_counts_per_bank = df.groupby('BANK_NAME')['SENTIMENT'].value_counts(normalize=True).unstack(fill_value=0)
print("\nSentiment Distribution per Bank (%):")
print((sentiment_counts_per_bank * 100).round(2))

# Plot overall rating distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='RATING', palette='viridis', order=sorted(df['RATING'].unique()))
plt.title('Overall Rating Distribution (1-5)')
plt.xlabel('Rating')
plt.ylabel('Number of Reviews')
plt.show()

# Plot overall sentiment distribution
plt.figure(figsize=(8, 8))
sentiment_counts = df['SENTIMENT'].value_counts(normalize=True) * 100
colors = {'positive': 'lightgreen', 'negative': 'lightcoral', 'neutral': 'lightblue'}
sentiment_counts = sentiment_counts.reindex(['positive', 'negative', 'neutral'], fill_value=0)
sentiment_colors = [colors[s] for s in sentiment_counts.index]

sentiment_counts.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    pctdistance=0.85,
    colors=sentiment_colors
)
plt.title('Overall Sentiment Distribution Across All Banks')
plt.ylabel('')
plt.show()

# Plot average rating per bank
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_rating_per_bank.index, y=avg_rating_per_bank.values, palette='coolwarm')
plt.title('Average Rating per Bank')
plt.xlabel('Bank Name')
plt.ylabel('Average Rating')
plt.ylim(1, 5)
plt.show()

# Monthly sentiment trend
df['review_month'] = df['REVIEW_DATE'].dt.to_period('M')
monthly_sentiment = df.dropna(subset=['review_month']).groupby(['review_month', 'BANK_NAME'])['SENTIMENT_SCORE'].mean().unstack()

plt.figure(figsize=(14, 7))
if not monthly_sentiment.empty:
    monthly_sentiment.plot(kind='line', marker='o', ax=plt.gca())
    plt.title('Monthly Average Sentiment Score Trend by Bank')
    plt.xlabel('Month')
    plt.ylabel('Average Sentiment Score')
    plt.xticks(rotation=45)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
else:
    print("Not enough data to plot monthly sentiment trend.")

# Word clouds for positive keywords
if drivers:
    positive_wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='Greens').generate_from_frequencies(drivers)
    plt.figure(figsize=(10, 5))
    plt.imshow(positive_wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Top Positive Keywords (Drivers)')
    plt.show()
else:
    print("No positive keywords identified.")

# Word clouds for negative keywords
if pain_points:
    negative_wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='Reds').generate_from_frequencies(pain_points)
    plt.figure(figsize=(10, 5))
    plt.imshow(negative_wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Top Negative Keywords (Pain Points)')
    plt.show()
else:
    print("No negative keywords identified.")
```
